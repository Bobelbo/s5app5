```python
import matplotlib.pyplot as plt
import numpy as np

a,b = -5, 5
N = [100,1000,10000]

avg_theo = (a + b) / 2
sigma_theo =  np.sqrt(np.square(b - a) / 12)

# N entre -5 et 5
pool = [(10 * np.random.random_sample(x)) - 5 for x in N]

avg = [np.mean(x) for x in pool]
med = [np.median(x) for x in pool]
err_quadra = [np.square(avg[i] - avg_theo) + np.square(med[i] - avg_theo) for i in range(len(N))]

print("Intervalle: [{}, {}]".format(a, b))
print("Moyenne Théorique:", avg_theo)
print("Ecart-type Théorique:", sigma_theo)
for i in range(len(N)):
    print(f"N={N[i]}\t Moyenne: {avg[i]}\t Médiane: {med[i]}\t Erreur quadratique: {err_quadra[i]}")
```

```python
def F_cfd(x):
    return [
        1 - (0.5 * (1 + np.sqrt(1 - np.exp(-np.square(i) * np.sqrt(np.pi/8)))))
        if i < 0 else
        0.5 * (1 + np.sqrt(1 - np.exp(-np.square(i) * np.sqrt(np.pi/8))))
        for i in x
    ]

def F_cfd_inv(p):
    return np.array([
        0
        if p == 1 or p == 0 else
        -np.sqrt(-np.log(1 - np.square(2 * p - 1))) / np.sqrt(np.pi/8) if p < 0.5 else
        np.sqrt(-np.log(1 - np.square(2 * (1 - p) - 1))) / np.sqrt(np.pi/8)
        for p in p
    ])

x = np.linspace(-4, 4, 1000)
xp = np.linspace(0, 1, 1000)
cdf = F_cfd(x)
inv = F_cfd_inv(xp)

N = [100,1000,10000]
pool = [avg_theo + sigma_theo * F_cfd_inv(np.random.rand(x)) for x in N]
avg = [np.mean(x) for x in pool]
med = [np.median(x) for x in pool]
err_quadra = [np.square(avg[i] - avg_theo) + np.square(med[i] - avg_theo) for i in range(len(N))]

plt.figure(figsize=(10, 6))
plt.plot(xp, inv, label='Inverse CDF', color='red')
plt.plot(x, cdf, label='CDF', color='blue')
plt.title('Fonction de répartition cumulative (CDF)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid()
plt.tight_layout()
plt.legend()

for i in range(len(N)):
    print(f"N={N[i]}\t Moyenne: {avg[i]}\t Médiane: {med[i]}\t Erreur quadratique: {err_quadra[i]}")

plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
plt.hist(pool[0], bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title(f'N={N[0]}')
plt.subplot(1, 3, 2)
plt.hist(pool[1], bins=30, alpha=0.7, color='orange', edgecolor='black')
plt.title(f'N={N[1]}')
plt.subplot(1, 3, 3)
plt.hist(pool[2], bins=30, alpha=0.7, color='green', edgecolor='black')
plt.title(f'N={N[2]}')
plt.suptitle('Histogrammes des échantillons normaux CDF inverse')
plt.tight_layout()
plt.show()
```

```python
R = 1
center = (0, 0, 0)
N = 100000

def distance(v1, v2):
    x0, y0, z0 = v1
    x1, y1, z1 = v2
    return np.sqrt(np.square(x1 - x0) + np.square(y1 - y0) + np.square(z1 - z0))

U1 = np.random.rand(N)
U2 = np.random.rand(N)
U3 = np.random.rand(N)

points = [(U1[i], U2[i], U3[i]) for i in range(N)]
rejets3D = [p for p in points if distance(p, center) <= R]
rejets2D = [(x,y,0) for (x,y,_) in points if distance((x,y,0), center) <= R]

aire2D_theo = np.pi * R**2
aire2D = 4 * (len(rejets2D) / N)

aire3D_theo = (4/3) * np.pi * R**3
aire3D = 8 * (len(rejets3D) / N)

print("Évaluation de l'aire du cercle (2D):", aire2D)
print("Aire théorique (2D):", aire2D_theo)
print("Évaluation de l'aire de la sphère (3D):", aire3D)
print("Aire théorique (3D):", aire3D_theo)
```
